```python
def bfs(root):
  queue = [root]
  visited = set()
  while queue:
    node = queue.pop(0)
    visited.add(node)
    print(node.value)
    #maze_search(node.value,'path')
    if node.left is not None and node.left not in visited:
      queue.append(node.left)
      #maze_search(node.value,'branch')
    if node.right is not None and node.right not in visited:
      queue.append(node.right)
      #maze_search(node.value,'branch')
    if node.value == 'N63' :
      break

# Perform BFS on the tree:
bfs(root)
#maze_display()
```

```python
def dfs(root):
  stack = [root]
  visited = set()

  while stack:
    node = stack.pop()
    visited.add(node)
    print(node.value)
    if node.right is not None and node.left not in visited:
      stack.append(node.right)
    if node.left is not None and node.right not in visited:
      stack.append(node.left)
    if node.value == 'N63' :
      break
  return visited

# Perform DFS on the tree:
dfs(root)
```
